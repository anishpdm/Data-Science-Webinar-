```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("covid_19_india.csv")
```

```python
df=pd.DataFrame(data)
```

```python
df.tail(n=3)
```

```python
df.head(n=3)
```

```python
df.describe()
```

```python
covid_latest_df=df[df['Date']=="21/05/20"]
```

```python
covid_latest_df=covid_latest_df.sort_values(by=['Confirmed'],ascending=False)
plt.figure(figsize=(20,20),dpi=100)


plt.bar(covid_latest_df['State/UnionTerritory'][:5],covid_latest_df.Deaths[:5],color="r")
plt.xlabel("State/UnionTerritory")
plt.ylabel("Deaths")
plt.title("Covid Graph")
plt.show()
```
